```python
def RDM(resp):
  """Compute the representational dissimilarity matrix (RDM)
  Args:
    resp (ndarray): S x N matrix with population responses to
      each stimulus in each row
  Returns:
    ndarray: S x S representational dissimilarity matrix
  """

  # z-score responses to each stimulus
  zresp = zscore(resp, axis=1)

  # Compute RDM
  RDM = 1 - (zresp @ zresp.T) / zresp.shape[1]

  return RDM


```

```python

def primary_matrix(change_data , trial_treshhold = 25 , min_index=38 , max_index = 62):
  matrix_data=np.array([])
  for cell_speci in change_data.cell_specimen_id.unique():
    row = np.array([])
    true_pure = change_data[(change_data.cell_specimen_id==cell_speci) & (change_data.rewarded==True)]
    false_pure = change_data[(change_data.cell_specimen_id==cell_speci) & (change_data.rewarded==False)]
    if len(true_pure) >= trial_treshhold and len(false_pure) >= trial_treshhold:

      traces_true = true_pure.trace.values[0:trial_treshhold]
      traces_false = false_pure.trace.values[0:trial_treshhold]
      for tr in traces_true:
        row = np.append(row,tr[min_index:max_index])

      for tr in traces_false:
        row = np.append(row,tr[min_index:max_index])

      matrix_data = np.append(matrix_data ,row )

  matrix_data=matrix_data.reshape(len(matrix_data)//(2*trial_treshhold*(max_index-min_index)),2*trial_treshhold*(max_index-min_index))
  print(matrix_data.shape[0])
  return matrix_data

def plot_RDM(change_data , trial_treshhold = 25 , min_index=38 , max_index = 62):
  pm = primary_matrix(change_data , trial_treshhold = trial_treshhold , min_index=min_index , max_index = max_index)
  rdm = RDM(pm.T)
  #print(pm)
  plot_corr_matrix(rdm)








```
